7. 정렬
- np.sort 함수
  - 원래 배열 데이터는 그대로 유지한 채 정렬된 배열이 복사본으로 반환

```python
x = np.array([4, 1, 5, 2, 3])
print(x, np.sort(x))
```

```python
#x.sort -> 원래 배열 자체를 정렬
x.sort()
print(x)
```

```python
#np.argsort 함수 -> 정렬된 요소의 인덱스를 반환
x = np.array([4, 1, 5, 2, 3])
i = np.argsort(x)
print(i)
```

```python
#x_reverse -> 내림차순 정렬
x = np.array([4, 1, 5, 2, 3])
x_reverse = np.sort(x)[::-1]
print(x_reverse)
```

- 다차원 배열 정렬

```python
#axis 인자 -> axis = 0일때, 각 열을 기준으로 정렬
X = np.array([[2, 0, 6], [7, 3, 4], [1, 5, 3]])
np.sort(X, axis = 0)
```
